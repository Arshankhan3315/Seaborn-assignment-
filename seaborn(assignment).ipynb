# Question- 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

# Answer-1-Seaborn is a popular Python data visualization library that provides a high-level interface for drawing attractive and informative statistical graphics. Here are five common plots you can create using Seaborn along with their typical uses:

# Scatter Plot (scatterplot)
Use: Visualize the relationship between two numerical variables.
Example Uses: Explore correlations, patterns, or clusters in data points.
# Bar Plot (barplot)

Use: Display the distribution of categorical data.
Example Uses: Compare the frequency or count of different categories.
# Box Plot (boxplot)

Use: Show the distribution of numerical data, especially highlighting outliers and quartiles.
Example Uses: Identify skewness, variability, and outliers in a dataset.
# Violin Plot (violinplot)

Use: Combine a box plot and kernel density plot to display the distribution of numerical data for different categories.
Example Uses: Visualize the distribution of data and its probability density at various values.
# Heatmap (heatmap)

Use: Display a matrix dataset as a grid of colored squares, where colors represent the values in the matrix.
Example Uses: Explore correlations, patterns, or relationships in a dataset with multiple variables.

# Question- 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

# Answer-2-

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
fmri_data = sns.load_dataset("fmri")
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=fmri_data, x='timepoint', y='signal', hue='event', style='region', markers=True)
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

# Answer-3-

```python
titanic_data = sns.load_dataset("titanic")
```

```python
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')
```

```python
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

# Answer-4-

```python
diamonds_data = sns.load_dataset("diamonds")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True)
plt.title('Histogram of Diamond Price by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.legend(title='Cut')
plt.show()
```

# Question- 6: Use the "flights" dataset from seaborn to plot a heatmap.

# Answer-6-

```python
flights_data = sns.load_dataset("flights")
```

```python
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, cmap='Blues', annot=True, fmt='d')
plt.title('Passenger Count by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
